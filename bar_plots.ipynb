```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['10', '20', '30', '40', '50']
train = [13.1, 36.6, 3.9, 29.3, 0.7, 16.4]
valid = [11.8, 37.2, 4.2, 27.5, 0.7, 18.6]
test = [7.5, 47.3, 5.5, 27.2, 0.7, 11.7]
test2 = [13.3, 36.6, 3.87, 29.4, 0.56, 16.3]

x = np.arange(len(labels))  # the label locations
barWidth = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 8))
r1 = np.arange(len(train))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

plt.bar(r1, train, color='#a6cee3', width=barWidth, edgecolor='white', label='Train')
plt.bar(r2, valid, color='#1f78b4', width=barWidth, edgecolor='white', label='Valid')
plt.bar(r3, test, color='#b2df8a', width=barWidth, edgecolor='white', label='Test')
plt.bar(r4, test2, color='#33a02c', width=barWidth, edgecolor='white', label='Test 2004-2009')

# Add xticks on the middle of the group bars
axis_font = {'fontname':'Arial', 'size':'14'}
#plt.xlabel('Land Cover Class', **axis_font)
plt.ylabel('% of dataset', **axis_font)
plt.xticks([r + barWidth for r in range(len(train))], ['Artificial \nAreas', 'Grass & Herb \nVegetation', 'Brush \nVegetation', 'Tree \nVegetation', 'Bare Land', 'Watery Areas'],
           **axis_font)
plt.yticks(**axis_font)
 
# Create legend & Show graphic
plt.legend(prop={'size': 14})
plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap, ListedColormap
labels = ['10', '20', '30', '40', '50', '60']
user = [0.66, 0.70, 0.000164, 0.89, 0.02, 0.97]
producer = [0.54, 0.79, 0.17, 0.75, 0.61, 0.99]

x = np.arange(len(labels))  # the label locations
barWidth = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 8))
r1 = np.arange(len(user))
r2 = [x + barWidth for x in r1]

colors = ['#909896','#d1eb6b','#14d514', '#346d1b', '#d0aa4a', '#377bd4']
mycmap = ListedColormap(colors)

plt.bar(r1, user, color=colors, width=barWidth, edgecolor='white', label='Precision')
plt.bar(r2, producer, color=colors, width=barWidth, edgecolor='white', alpha=0.5, label='Recall')

# Add xticks on the middle of the group bars
axis_font = {'fontname':'Arial', 'size':'14'}
#plt.xlabel('Land Cover Class', **axis_font)
plt.ylabel('Score', **axis_font)
plt.xticks([r + barWidth for r in range(len(user))], ['Artificial Areas', 'Grass & Herb \nVegetation', 'Brush \nVegetation', 'Tree \nVegetation', 'Bare Land', 'Watery Areas'],
           **axis_font)
plt.yticks(**axis_font)
 
# Create legend & Show graphic
plt.legend(prop={'size': 14})
plt.show()
```
